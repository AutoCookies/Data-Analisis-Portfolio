```python
def count_employment (dataframe, text):
    return dataframe['Employment'].str.lower().str.contains(text).sum()

Employment_df_shorted = pd.DataFrame({
    'Status': ['Employed', 'Not Employed', 'Freelancer', 'Self-employed', 'Student', 'Looking for work', 'Retired'],
    'count': [
        count_employment(public_df, 'employed'),
        count_employment(public_df, 'not employed'),
        count_employment(public_df, 'freelancer'),
        count_employment(public_df, 'self-employed'),
        count_employment(public_df, 'student'),
        count_employment(public_df, 'looking for work'),
        count_employment(public_df, 'retired')
        ]
    }
)

Employment_df_shorted
```

## Function

Bar chart function

```python
def bar_chart (dataframe, title_, xlabel, ylabel, colors, figsize_ = (18, 10), ):
    ax = dataframe.plot(
        kind = 'bar',
        figsize = figsize_,
        title= title_,
        color= colors
    )
    
    annotate_chart(ax)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    
def annotate_chart (ax, pad= 1, color='white', textsize = 14):
    for i in ax.patches:
    # Calculate Annotation
        value = str(round(i.get_height(), 1))
        if value == '0.0':
            continue
        
        annotation_x = i.get_x() + i.get_width() / 2
        annotation_y = i.get_y() + i.get_height() / 2
        
        ax.annotate(
            value,
            (annotation_x, annotation_y),
            color=color,
            size=textsize,
            ha='center',  # horizontal alignment
            va='center'   # vertical alignment
        )
```

Line Chart function

```python
def line_chart (dataframe, title_, xlabel, ylabel, colors, figsize_ = (18, 10), ):
    ax = dataframe.plot(
        kind = 'line',
        figsize = figsize_,
        title= title_,
        color= colors
    )
    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

Stacked Bar Chart

```python
def plot_stacked_bar(dataframe, title_, xlabel, ylabel, size_=(20, 10), rot_=0, legend_='upper right'):

    ax = dataframe.plot(
        kind='bar',
        stacked=True,
        figsize=size_,
        rot=rot_,
        xlabel = xlabel,
        ylabel = ylabel,
        title=title_
    )
    
    annotate_chart(ax)
    plt.show()
```
